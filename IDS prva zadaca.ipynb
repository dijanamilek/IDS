8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
students.head(10)
```

```python
students['EthnicGroup'].fillna(students['EthnicGroup'].mode()[0], inplace=True)
students['ParentMaritalStatus'].fillna(students['ParentMaritalStatus'].mode()[0], inplace=True)
students['WklyStudyHours'].fillna(students['WklyStudyHours'].mode()[0], inplace=True)
students[['ParentEduc']] = students[['ParentEduc']].fillna('no formal education')
students[['TestPrep']] = students[['TestPrep']].fillna('none')
# remove records with missing values of feature

students.loc[students.NrSiblings.isna(),'NrSiblings'] = students.loc[:,'NrSiblings'].median() 
students[['TransportMeans']] = students[['TransportMeans']].fillna('walking')
students.isna().sum()
#df.fillna(df.mode().iloc[0])
#df.mode = most common value
#df[['Courses','Fee']] = df[['Courses','Fee']]. fillna('') --> replace with string
```

```python
students_cleaned = students.copy()
students_cleaned=students.loc[students.PracticeSport.notnull(), :]
students_cleaned=students_cleaned.loc[students.IsFirstChild.notnull(), :].reset_index(drop = True)

students_cleaned.isna().sum()
#students = pd.concat((students, students_tmp), axis=1)
```

9. Are there inconsistencies in the IsFirstChild column? Replace the inconsistent entries so they match the rest of the values. 

```python
#ili: students['IsFirstChild'].drop_duplicates()
FirstChild = students_cleaned['IsFirstChild'].unique()
FirstChild
pd.crosstab(students_cleaned['IsFirstChild'], students_cleaned['Gender'])
```

```python
students_cleaned['IsFirstChild'] = students_cleaned['IsFirstChild'].replace(['Yes','es','si','ye','yeah','yeah boy','ys'], 'yes')

students_cleaned['IsFirstChild'] = students_cleaned['IsFirstChild'].replace(['nein','noo','nope'], 'no')

pd.crosstab(students_cleaned['IsFirstChild'], students_cleaned['Gender'])
```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
students_tmp=students_cleaned.copy()
# we define the encoder
encoder = OneHotEncoder(handle_unknown="ignore")
encoder.fit(students_cleaned[['Gender','LunchType']])
# we transform the data
students_tmp = encoder.transform(students_cleaned[['Gender','LunchType']])
students_tmp = pd.DataFrame(students_tmp.toarray(), columns=encoder.get_feature_names_out(['Gender','LunchType']))
students_tmp
```

```python
students_cleaned= pd.concat((students_cleaned, students_tmp), axis=1) #dodajemo u dataframe
students_cleaned
```

```python
# initialization of the encoder
le = LabelEncoder()
students_cleaned["EthnicGroup"] =le.fit_transform(students_cleaned["EthnicGroup"])
students_cleaned["ParentEduc"] =le.fit_transform(students_cleaned["ParentEduc"])
students_cleaned["TestPrep"] =le.fit_transform(students_cleaned["TestPrep"])
students_cleaned["ParentMaritalStatus"] =le.fit_transform(students_cleaned["ParentMaritalStatus"])
students_cleaned["PracticeSport"] =le.fit_transform(students_cleaned["PracticeSport"])
students_cleaned["IsFirstChild"] =le.fit_transform(students_cleaned["IsFirstChild"])
students_cleaned["TransportMeans"] =le.fit_transform(students_cleaned["TransportMeans"])
students_cleaned["WklyStudyHours"] =le.fit_transform(students_cleaned["WklyStudyHours"])



#le.fit(students_cleaned.loc[:,["EthnicGroup", "ParentEduc", "TestPrep", "ParentMaritalStatus", "PracticeSport", "IsFirstChild", "TransportMeans", "WklyStudyHours"]])

#students_cleaned.loc[:,["EthnicGroup", "ParentEduc", "TestPrep", "ParentMaritalStatus", "PracticeSport", "IsFirstChild", "TransportMeans", "WklyStudyHours"]
#] = le.transform(students_cleaned.loc[:,["EthnicGroup", "ParentEduc", "TestPrep", "ParentMaritalStatus", "PracticeSport", "IsFirstChild", "TransportMeans", "WklyStudyHours"]
#])
students_cleaned
```
